```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from IPython.core.pylabtools import figsize
```

```python
figsize(10, 10)
```

```python
df = pd.read_csv("data/final/segmented.csv")
```

```python
df.columns
```

```python
def create_box_plot(df: pd.DataFrame, y: str):
    sorted_clusters = (
        df.groupby(["clusters"]).agg({y: "median"}).sort_values(by=y).index
    )
    sns.boxplot(data=df, x="clusters", y=y, order=sorted_clusters)
    plt.plot()
```

```python
columns = ["Income", "NumWebVisitsMonth", "age", "total_purchases", "family_size"]
```

```python
# Parameters
columns = [
    "Income",
    "Recency",
    "NumWebVisitsMonth",
    "NumDealsPurchases",
    "NumWebPurchases",
    "NumCatalogPurchases",
    "NumStorePurchases",
    "Complain",
    "Response",
    "age",
    "enrollment_years",
    "family_size",
]
```

```python
for col in columns:
    create_box_plot(df, col)
    plt.plot()
    plt.show()
```
